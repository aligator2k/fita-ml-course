## Modeļu un pieeju salīdzinājums

**Kad lietot klasifikāciju?**
Klasifikāciju lieto, kad atbilde ir kategorija. Piemēram, medicīnā: vai pacientam ir diagnoze "diabēts" (Jā/Nē), balstoties uz asins analīžu rādītājiem. Target ir diskrēts — 0 vai 1.

**Kad lietot regresiju?**
Regresiju lieto, kad atbilde ir skaitlis. Piemēram, nekustamajā īpašumā: cik EUR vērts ir konkrēts dzīvoklis, balstoties uz platību, stāvu un atrašanās vietu. Target ir nepārtraukts — jebkurš skaitlis.

**Kad lietot klasterizāciju?**
Klasterizāciju lieto, kad nav zināmā target kolonnas un mēs meklējam dabiskas grupas. Piemēram, mārketingā: kādas klientu grupas pastāv, balstoties uz pirkumu biežumu un vidējo čeka summu — bez iepriekšējas zināšanas, cik grupu ir.

**Kāpēc Pipeline un Cross-validation ir nepieciešami?**
Pipeline novērš data leakage — scaler tiek fit tikai uz treniņa datiem katrā fold. Cross-validation dod uzticamu modeļa novērtējumu, jo vidējo no 5 neatkarīgiem eksperimentiem, nevis paļaujas uz vienu nejaušu split. Bez tiem rezultāti var būt optimistiski neprecīzi un nereprezentabli.
